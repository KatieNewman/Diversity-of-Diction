```python
import re
import numpy as np
import pandas as pd
import nltk
import requests
from bs4 import BeautifulSoup
import urllib.request
```

```python
with open('DBarstow1.txt', 'r') as file:
    journalist1_article1_text = file.read().replace('\n', ' ')
#     text1 = file.read()
```

```python
journalist1_article1_text
```

```python
journalist1_article1_text_tokens = re.findall('\w+', journalist1_article1_text)
```

```python
journalist1_article1_text_tokens[:10]
```

```python
journalist1_article1_text_words_regardless_of_case = []

for word in journalist1_article1_text_tokens:
    journalist1_article1_text_words_regardless_of_case.append(word.lower())


# Print several items from list as sanity check
journalist1_article1_text_words_regardless_of_case[:10]
```

```python
len(journalist1_article1_text_words_regardless_of_case)
```

```python
sw = nltk.corpus.stopwords.words('english')
sw[:5]
```

```python
journalist1_article1_text_non_stop_words = []

# Add to words_ns all words that are in words but not in sw
for word in journalist1_article1_text_words_regardless_of_case:
    if word not in sw:
        journalist1_article1_text_non_stop_words.append(word)

# Print several list items as sanity check
journalist1_article1_text_non_stop_words[:5]
```

```python
len(journalist1_article1_text_non_stop_words)
```

```python
# lister = ["bank", "buck", "bank", "bork", "bank", "buck", "bike", "bank", "bork", "buck"]

journalist1_article1_text_dictionary_of_words_used = {}

for x in journalist1_article1_text_non_stop_words:
   if x in journalist1_article1_text_dictionary_of_words_used:
       journalist1_article1_text_dictionary_of_words_used[x] = journalist1_article1_text_dictionary_of_words_used[x] + 1
   else:
       journalist1_article1_text_dictionary_of_words_used[x] = 1
# print(journalist1_article1_text_dictionary_of_words_used)
```

```python
journalist1_article1_text_words_and_freq_df = pd.DataFrame(list(journalist1_article1_text_dictionary_of_words_used.items()))
```

```python
journalist1_article1_text_words_and_freq_df
```

```python
journalist1_article1_text_words_and_freq_df['title'] = "DBarstow1"
```

```python
journalist1_article1_text_words_and_freq_df
```

```python
journalist1_article1_text_words_and_freq_df = journalist1_article1_text_words_and_freq_df.rename(index=str, columns={0: "word"})
```

```python
journalist1_article1_text_words_and_freq_df = journalist1_article1_text_words_and_freq_df.rename(index=str, columns={1: "frequency"})
```

```python
journalist1_article1_text_words_and_freq_df
```

```python
all_columns_partially_filled_csv = "all_columns_partially_filled_df.csv"

all_columns_partially_filled_df = pd.read_csv(all_columns_partially_filled_csv)
```

```python
all_columns_partially_filled_df
```

```python
all_columns_partially_filled_df = pd.merge(all_columns_partially_filled_df, journalist1_article1_text_words_and_freq_df, on=["title","word","frequency"], how="outer")
```
